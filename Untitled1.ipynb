```python
import pandas as pd
import seaborn as sns
data = pd.read_csv('HPI.csv')
print(data.head())
```

```python
sns.histplot(data['living_area_renov'])
```

```python
sns.countplot(x=data['Date'])
```

```python
import matplotlib.pyplot as plt
x = data['Date'].value_counts()
plt.pie(x.values,
        labels=x.index,
        autopct='%1.1f%%')
plt.show()
```

```python
#Bivariate Analysis
plt.figure(figsize=(15, 5))
sns.barplot(x=data['Date'], y=data['Price'])
plt.xticks(rotation='90')
```

```python
sns.scatterplot(x=data['Date'],
                y=data['Price'])
```

```python
sns.countplot(data['Price'],hue=data['Price'])
```

```python
#Multivariate Analysis
from sklearn import datasets, decomposition
iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```
